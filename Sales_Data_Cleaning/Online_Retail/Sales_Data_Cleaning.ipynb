## Step 5: Check Missing Values

This step identifies missing values in each column and calculates their percentage.

```python
df.isnull().sum()
```

```python
(df.isnull().sum() / len(df)) * 100
```

## Step 6: Check and Remove Duplicate Records

Identify duplicate records and remove them.

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
df.duplicated().sum()
```

## Step 7: Handle Missing Values

Fill missing values in the Description and CustomerID columns.

```python
df["Description"] = df["Description"].fillna("Unknown")
```

```python
df["CustomerID"] = df["CustomerID"].fillna(0)
```

```python
df.isnull().sum()
```

## Step 8: Clean Column Names

Remove extra spaces from column names.

```python
df.columns = df.columns.str.strip()
```

```python
df.columns
```

## Step 9: Convert Date Column

Convert the InvoiceDate column into datetime format.

```python
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"])
```

```python
df.dtypes
```

## Step 10: Count Unique Values

Display the number of unique values in each column.

```python
df.nunique()
```

## Step 11: Display the Cleaned Dataset

Preview the cleaned dataset.

```python
df.head()
```

## Step 12: Save the Cleaned Dataset

Save the cleaned dataset as a new CSV file.

```python
df.to_csv("Cleaned_Online_Retail.csv", index=False)
```
