## Step 12: Remove Duplicate Records

Remove duplicate rows from the dataset.

```python
df=df.drop_duplicates()

df.duplicated().sum()
```

## Step 13: Handle Missing Values

Fill missing values in the Item_Weight and Outlet_Size columns.

```python
df["Item_Weight"] = df["Item_Weight"].fillna(df["Item_Weight"].mean())

df["Outlet_Size"] = df["Outlet_Size"].fillna(df["Outlet_Size"].mode()[0])
```

```python
df.isnull().sum()
```

## Step 14: Standardize Item_Fat_Content Values

Replace inconsistent category names with standard values.

```python
df["Item_Fat_Content"] = df["Item_Fat_Content"].replace({
    "LF": "Low Fat",
    "low fat": "Low Fat",
    "reg": "Regular"
})
```

```python
df["Item_Fat_Content"].value_counts()
```

## Step 15: Save the Cleaned Dataset

Save the cleaned dataset as a new CSV file.

```python
df.to_csv("Cleaned_BigMart_Sales.csv", index=False)
```

## Visualization 1: Distribution of Item Outlet Sales (Histogram)

```python
plt.figure(figsize=(8,5))

plt.hist(df['Item_Outlet_Sales'], bins=20, edgecolor='black')

plt.title('Distribution of Item Outlet Sales')
plt.xlabel('Item Outlet Sales')
plt.ylabel('Frequency')

plt.show()
```

## Visualization 2: Average Sales by Item Type (Bar Chart)

```python
plt.figure(figsize=(12,6))

df.groupby('Item_Type')['Item_Outlet_Sales'].mean().sort_values().plot(kind='bar')

plt.title('Average Sales by Item Type')
plt.xlabel('Item Type')
plt.ylabel('Average Sales')

plt.show()
```

## Visualization 3: Outlet Size Distribution (Pie Chart)

```python
plt.figure(figsize=(7,7))

df['Outlet_Size'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    startangle=90
)

plt.title('Outlet Size Distribution')
plt.ylabel('')

plt.show()
```

## Visualization 4: Item MRP vs Item Outlet Sales (Scatter Plot)

```python
plt.figure(figsize=(8,5))

plt.scatter(
    df['Item_MRP'],
    df['Item_Outlet_Sales'],
    alpha=0.5
)

plt.title('Item MRP vs Item Outlet Sales')
plt.xlabel('Item MRP')
plt.ylabel('Item Outlet Sales')

plt.show()
```

## Visualization 5: Correlation Heatmap

```python
import seaborn as sns

plt.figure(figsize=(8,6))

sns.heatmap(
    df.corr(numeric_only=True),
    annot=True,
    cmap='coolwarm'
)

plt.title('Correlation Heatmap')

plt.show()
```
